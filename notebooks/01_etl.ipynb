Aquí leeremos la tabla y exáminaremos los datos


```python
import pandas as pd

raw_path = "../data/raw/stock_senti_analysis.csv"

# Leer CSV con codificación latin1
df = pd.read_csv(raw_path, encoding='latin1')

print(df.head())
print(df.info())
```

```python
# 1. Convertir columna 'Date' a datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# 2. Eliminar filas con fecha inválida
df = df.dropna(subset=['Date'])

# 3. Eliminar duplicados
df = df.drop_duplicates()

# 4. Rellenar valores nulos en titulares con cadena vacía
for col in df.columns[2:]:  # Top1 a Top25
    df[col] = df[col].fillna("")

# 5. Normalizar texto: minúsculas y eliminar espacios extras
for col in df.columns[2:]:
    df[col] = df[col].str.lower().str.strip()

# 6. Revisar cambios
print(df.info())
print(df.head())
```

```python
# Guardar CSV limpio
df.to_csv("../data/processed/DJIA_news_clean.csv", index=False)

# Con pyarrow
df.to_parquet("../data/processed/DJIA_news_clean.parquet", index=False, engine='pyarrow')
# Convertir todas las columnas de tipo object a string
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str)
# Guardar en SQLite
import sqlite3
conn = sqlite3.connect("../data/processed/djia_news.db")
df.to_sql("djia_news", conn, if_exists="replace", index=False)
conn.close()
```
